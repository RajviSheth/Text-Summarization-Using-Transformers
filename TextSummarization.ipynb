```python
from transformers import pipeline
```

```python
summarizer = pipeline('summarization')
```

```python
article = """
INOWEAVE, an Original Design Manufacturing Company established in 1987, with a focus on providing customized embedded electronics solutions. We are a concept-to-production company which offers services for design development and manufacturing of customized industrial electronics controls with firmware (Embedded Solutions) to satisfy the specific needs of our customers.

Our vision is to provide reliable, cost-effective and exclusively designed embedded solutions to machine manufacturers for any industry. We have a team of qualified and experienced engineers to develop high-performance products with complex logic along with our well-equipped manufacturing facility for providing total cost-effective solutions.

industrial electronics controls

Customized Industrial Electronics Controls
Manufacturing in Surat
The Electronics Industry is among the fastest-growing engineering branches. It has evolved much faster than mankind. Adopting and evolving along with the latest technology requires a team of highly skilled engineers, latest development tools & software and modern efficient manufacturing equipment. All these require constant capital intensive upgrades as technology gets outdated every few years.
We understand that industrial electronics controls contributes a part of your entire product but at Inoweave that’s 100% of our activity. So, we can afford to provide quick and economical embedded solutions for your products.

“We develop embedded solutions exclusively for a customer and do NOT accept similar proposals from other OEMs  to enhance product secrecy for our customers”


"""
```

```python
# the do_sample is for greedy decoder which states which word to return next and which word makes the most sense
summary = summarizer(article, max_length=130, min_length=30, do_sample=False)
```
